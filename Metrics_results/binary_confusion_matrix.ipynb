```python
import pandas as pd 
import seaborn as sns
import pickle as pkl
import glob
import matplotlib.pyplot as plt 

from matplotlib import rcParams
rcParams['font.family'] = 'Arial'
```

# HS Classifier

```python

directory = f'../Metrics_results/distilbert-base-cased/test/blacklist-jigsaw-DistilBERT-binary-FocalLoss_test.pkl'
file_list, data = glob.glob(directory), []

for pkl_file in file_list:
    with open(pkl_file, 'rb') as file:
        data.append(pkl.load(file))
data 
```

```python
data[0][0]['all_metrics']
```

```python
conf_matrix = data[0][0]['all_metrics']['confusion_matrix'].cpu().numpy()

# Plot confusion matrix
plt.figure(figsize=(3, 3))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'], cbar=False)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Blacklist CLS Confusion Matrix')
plt.show()
```
