```python
# GET all the comments which are not flagged as hate speech
condition1 = jigsaw_train['toxicity'] == 0
condition2 = jigsaw_train['obscene'] == 0
condition3 = jigsaw_train['sexual_explicit'] == 0
condition4 = jigsaw_train['identity_attack'] == 0
condition5 = jigsaw_train['threat'] == 0
condition6 = jigsaw_train['insult'] == 0
train_not_offensive = jigsaw_train[condition1 & condition2 & condition3 & condition4 & condition5 & condition6]

#Get all the comments that are hateful excluding the ones that are not

train_without_not_offensive = jigsaw_train[jigsaw_train[['toxicity','obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']].sum(axis=1) != 0]
len(train_not_offensive), len(train_without_not_offensive)
```

```python
not_offensive_train_scaled = train_not_offensive.sample(frac=0.3, random_state=42)
```

```python
len(not_offensive_train_scaled)
```

```python
training_samples = pd.concat([train_without_not_offensive, not_offensive_train_scaled], axis=0)
training_samples = training_samples.sample(frac=1, random_state=42)
training_samples
```

```python
from sklearn.model_selection import train_test_split


train, validation, y_train, y_val = train_test_split(training_samples, training_samples[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']], test_size=.2, shuffle=True, random_state=42)
```

```python
train.to_csv('../Data/jigsaw.train.csv')
validation.to_csv('../Data/jigsaw.validation.csv')
```
