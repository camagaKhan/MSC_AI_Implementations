```python
import pandas as pd
from transformers import AutoTokenizer
```

```python
train, validation, test = pd.read_csv('../Data/jigsaw.train.csv'), pd.read_csv('../Data/jigsaw.validation.csv'), pd.read_csv('../Data/jigsaw.test.csv')
```

```python
import numpy as np
```

```python
train['not_offensive'] = (train[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']].sum(axis=1) == 0).astype(np.float32)
test['not_offensive'] = (test[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']].sum(axis=1) == 0).astype(np.float32)
validation['not_offensive'] = (validation[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']].sum(axis=1) == 0).astype(np.float32)

for label in ['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']:
    train[label] = train.apply(lambda row: 0 if row['not_offensive'] == 1 else row[label], axis=1)
    validation[label] = validation.apply(lambda row: 0 if row['not_offensive'] == 1 else row[label], axis=1)
    test[label] = test.apply(lambda row: 0 if row['not_offensive'] == 1 else row[label], axis=1)
```

```python
train
```

```python
train.to_csv('../Data/jigsaw.train.csv')
validation.to_csv('../Data/jigsaw.validation.csv')
test.to_csv('../Data/jigsaw.test.csv')
```
