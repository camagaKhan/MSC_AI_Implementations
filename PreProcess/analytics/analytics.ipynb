# Custom Dataset

```python
annotator_choice = pd.read_csv('../../Data/annotator_choice.csv')
annotator_choice
```

```python
annotator_choice[[ 'Toxicity', 'Obscene',	'Sexual_Explicit',	'Identity_Attack',	'Insult',	'Threat' ]] = annotator_choice[[ 'Toxicity', 'Obscene',	'Sexual_Explicit',	'Identity_Attack',	'Insult',	'Threat' ]].fillna(0.0)
annotator_choice
```

```python
# Pivot the dataset for each label separately, with one column per annotator
toxicity_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Toxicity')
obscene_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Obscene')
insult_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Insult')
sexual_explicit_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Sexual_Explicit')
identity_attack_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Identity_Attack')
threat_pivot = annotator_choice.pivot(index='CommentID', columns='AnnotatorID', values='Threat')

# Rename columns to reflect the annotators
toxicity_pivot.columns = [f'Annotator {col} (Toxicity)' for col in toxicity_pivot.columns]
obscene_pivot.columns = [f'Annotator {col} (Obscene)' for col in obscene_pivot.columns]
insult_pivot.columns = [f'Annotator {col} (Insult)' for col in insult_pivot.columns]
sexual_explicit_pivot.columns = [f'Annotator {col} (Sexual_Explicit)' for col in sexual_explicit_pivot.columns]
identity_attack_pivot.columns = [f'Annotator {col} (Identity_Attack)' for col in identity_attack_pivot.columns]
threat_pivot.columns = [f'Annotator {col} (Threat)' for col in threat_pivot.columns]

# Merge the pivoted tables into a single DataFrame
result = toxicity_pivot.join([obscene_pivot, insult_pivot, sexual_explicit_pivot, identity_attack_pivot, threat_pivot])

#result.fillna(0.0, inplace=True)
#result = result.dropna()
# result.dropna()
# Reset index to make CommentID a column
result.reset_index(inplace=True)
```

```python
result
```

```python
# Extract columns related to 'Toxicity'
toxicity_columns = [col for col in result.columns if 'Toxicity' in col]
toxicity_data = result[toxicity_columns].values

# Extract columns related to 'Obscene'
obscene_columns = [col for col in result.columns if 'Obscene' in col]
obscene_data = result[obscene_columns].values

# Extract columns related to 'Insult'
insult_columns = [col for col in result.columns if 'Insult' in col]
insult_data = result[insult_columns].values

# Extract columns related to 'Sexual_Explicit'
sexual_explicit_columns = [col for col in result.columns if 'Sexual_Explicit' in col]
sexual_explicit_data = result[sexual_explicit_columns].values

# Extract columns related to 'Identity_Attack'
identity_attack_columns = [col for col in result.columns if 'Identity_Attack' in col]
identity_attack_data = result[identity_attack_columns].values

# Extract columns related to 'Threat'
threat_columns = [col for col in result.columns if 'Threat' in col]
threat_data = result[threat_columns].values
```

```python
import krippendorff
```

```python
# Calculate Krippendorff's alpha for 'Toxicity'
toxicity_alpha = krippendorff.alpha(reliability_data=toxicity_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Toxicity: {toxicity_alpha}")

# Calculate Krippendorff's alpha for 'Obscene'
obscene_alpha = krippendorff.alpha(reliability_data=obscene_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Obscene: {obscene_alpha}")

# Calculate Krippendorff's alpha for 'Insult'
insult_alpha = krippendorff.alpha(reliability_data=insult_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Insult: {insult_alpha}")

# Calculate Krippendorff's alpha for 'Sexual_Explicit'
sexual_explicit_alpha = krippendorff.alpha(reliability_data=sexual_explicit_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Sexual Explicit: {sexual_explicit_alpha}")

# Calculate Krippendorff's alpha for 'Identity_Attack'
identity_attack_alpha = krippendorff.alpha(reliability_data=identity_attack_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Identity Attack: {identity_attack_alpha}")

# Calculate Krippendorff's alpha for 'Threat'
threat_alpha = krippendorff.alpha(reliability_data=threat_data, level_of_measurement='nominal')
print(f"Krippendorff's alpha for Threat: {threat_alpha}")
```
