```python
dataset = dataset.drop(columns=['id', 'severe_toxic'])
dataset
```

```python
dataset = dataset.rename(columns={'toxic': 'toxicity', 'identity_hate': 'identity_attack' })
dataset
```

```python
dataset = dataset[['comment_text', 'toxicity',	'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']]
dataset
```

```python
import re 
from cleantext import clean

https_url_pattern = re.compile(r'https?://\S+|www\.\S+') 
remove_tags = re.compile('<.*?>')

def remove_urls_and_htmlTags(text, replacement_text=''):
    # Define a regex pattern to match URLs
    text_without_urls = https_url_pattern.sub(replacement_text, text)
    result = remove_emojis(text_without_urls)
    return re.sub(remove_tags, replacement_text, result)

def remove_emojis (text):
    text = text.replace('"', '')
    if text != '':        
        text = clean(text,
                    # no_emoji=True, 
                    extra_spaces=True,
                    lowercase=False) 
    cleaned_text = re.sub(r'&#[0-9]+;', '', text)
    cleaned_text = text.replace('\n', ' ')
    cleaned_text = re.sub(r'@\w+', '[USER]', text)
    return cleaned_text
```

```python
dataset['comment_text'] = dataset['comment_text'].apply(lambda x : remove_urls_and_htmlTags(x))
```

```python
dataset[dataset['comment_text'] == '']
```

```python
dataset = dataset[dataset['comment_text'].str.strip().astype(bool)]
```

```python
dataset.loc[:, 'comment_text'] = dataset['comment_text'].str.replace('\n', '', regex=False)
```

```python
dataset.to_csv('../Data/wiki_corpus_cleaned.csv')
```

```python
from sklearn.model_selection import train_test_split


train, validation, y_train, y_val = train_test_split(dataset, dataset[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']], test_size=.2, shuffle=True, random_state=42)
```

```python
train, test, y_train, y_val = train_test_split(train, train[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']], test_size=.2, shuffle=True, random_state=42)
```
