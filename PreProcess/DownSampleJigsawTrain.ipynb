```python
import pandas as pd
```

```python
train, validation = pd.read_csv('../Data/jigsaw.train.multi-label_mixed.csv'), pd.read_csv('../Data/jigsaw.validation.multi-label_mixed.csv')
train
```

```python
condition1 = train['toxicity'] == 0.0
condition2 = train['obscene'] == 0.0
condition3 = train['sexual_explicit'] == 0.0
condition4 = train['identity_attack'] == 0.0
condition5 = train['threat'] == 0.0
condition6 = train['insult'] == 0.0
train_w_not_offensive = train[condition1 & condition2 & condition3 & condition4 & condition5 & condition6]
```

```python
not_offensive_w_train_scaled = train_w_not_offensive.sample(frac=0.15, random_state=42).reset_index(drop=True)
```

```python
train_without_w_not_offensive = train[train[['toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat']].sum(axis=1) != 0]
len(train_without_w_not_offensive)
```

```python
train = pd.concat([not_offensive_w_train_scaled, train_without_w_not_offensive], ignore_index=True)
train
```

```python
train = train.sample(frac=1).reset_index(drop=True)
```
