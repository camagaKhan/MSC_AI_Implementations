I have a major issue here. Dataset is still highly imbalanced

```python
labels = ['toxicity',	'obscene',	'sexual_explicit',	'identity_attack',	'insult', 'threat']
counts_train = []

for label in labels:
    counts_train.append(train[label].value_counts().get(1., 0))
    
print(counts_train)
```

So...

```python
toxic_indexes, insult_indexes = train.index[train['toxicity'] == 1.].to_list(), train.index[train['insult'] == 1.].to_list()
print("Toxic indexes:", toxic_indexes)
print("Insult indexes:", insult_indexes)

# new_training_dataset
```

As you can see, indexes appear in both subsets. So, this is what I am going to do:

```python
unique_indexes = list(set(toxic_indexes + insult_indexes))
new_training_dataset_1 = train.loc[unique_indexes]
new_training_dataset_1
```

```python
new_training_dataset_1['toxicity'].value_counts().get(1., 0), new_training_dataset_1['insult'].value_counts().get(1., 0)
```

```python
obscene_indexes, se_indexes, ia_indexes, threat_indexes = train.index[train['obscene'] == 1.].to_list(), train.index[train['sexual_explicit'] == 1.].to_list(), train.index[train['identity_attack'] == 1.].to_list(), train.index[train['threat'] == 1.].to_list()
print("Obscene indexes:", obscene_indexes)
print("SE indexes:", se_indexes)
print("IA indexes:", ia_indexes)
print("threat indexes:", threat_indexes)
```

```python
unique_indexes = list(set(obscene_indexes + se_indexes + ia_indexes + threat_indexes))
new_training_dataset_2 = train.loc[unique_indexes]
new_training_dataset_2
```

```python
new_training_dataset_1 = new_training_dataset_1.drop(columns=['Unnamed: 0.1','Unnamed: 0', 'obscene', 'sexual_explicit', 'identity_attack', 'threat'	])
new_training_dataset_1
```

```python
new_training_dataset_2 = new_training_dataset_2.drop(columns=['Unnamed: 0.1','Unnamed: 0', 'toxicity', 'insult'	])
new_training_dataset_2
```

```python
new_training_dataset_1.to_csv('../Data/jigsaw.train.multi-label.majority.csv', index=False)
new_training_dataset_2.to_csv('../Data/jigsaw.train.multi-label.minority.csv', index=False)
```
