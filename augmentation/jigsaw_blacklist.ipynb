```python
import pandas as pd
import os
```

```python
all_dataset = pd.read_csv('../Data/jigsaw.train.csv')
all_dataset.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
all_dataset
```

```python
# Define the maximum number of rows per CSV file (for Google Sheets)
max_rows_per_file = 32_900  # Adjust based on your needs or Google Sheets limits

# Create a directory for the smaller CSV files
output_dir = 'split_csvs'
os.makedirs(output_dir, exist_ok=True)

# Split the records into smaller CSV files
for i in range(0, len(all_dataset), max_rows_per_file):
    subset_df = all_dataset.iloc[i:i + max_rows_per_file]
    subset_filename = os.path.join(output_dir, f'split_train_{i // max_rows_per_file + 1}.csv')
    subset_df.to_csv(subset_filename, index=False)

print(f"Split files saved in the directory: {output_dir}")
```

```python
val, test = pd.read_csv('../Data/jigsaw.validation.csv'), pd.read_csv('../Data/jigsaw.test.csv')

# Create a directory for the smaller CSV files
output_dir = 'split_csvs'
os.makedirs(output_dir, exist_ok=True)

# Split the records into smaller CSV files
for i in range(0, len(val), max_rows_per_file):
    subset_val_df = val.iloc[i:i + max_rows_per_file]
    subset_val_filename = os.path.join(output_dir, f'split_val_{i // max_rows_per_file + 1}.csv')
    subset_val_df.to_csv(subset_val_filename, index=False)

for i in range(0, len(test), max_rows_per_file):    
    subset_test_df = test.iloc[i:i + max_rows_per_file]
    subset_test_filename = os.path.join(output_dir, f'split_test_{i // max_rows_per_file + 1}.csv')
    subset_test_df.to_csv(subset_test_filename, index=False)

print(f"Split files saved in the directory: {output_dir}")

#'../Data/jigsaw.train.csv')
#all_dataset.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
```
