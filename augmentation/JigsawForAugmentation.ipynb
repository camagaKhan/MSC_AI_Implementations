````python
import pandas as pd 
````

````python
dataset = pd.read_csv("../Data/jigsaw.15.train.multi-label.csv", usecols=lambda column: column not in ['Unnamed: 0'])
dataset
````

````python
df = dataset
df = df[['toxicity',	'obscene',	'sexual_explicit',	'identity_attack',	'insult',	'threat']]
counts = df.sum().astype(int)
print(counts)
````

````python
no_toxicity = dataset[dataset['toxicity'] == 0]
df_filtered = no_toxicity[no_toxicity[['obscene',	'sexual_explicit',	'identity_attack',	'insult',	'threat']].any(axis=1)]
df_filtered
````

````python
df_filtered.to_csv('dataset.without.toxicity.csv')
````

````python
df = df_filtered
df = df[['toxicity', 'obscene',	'sexual_explicit', 'identity_attack', 'insult', 'threat']]
counts = df.sum().astype(int)
print(counts)
````
