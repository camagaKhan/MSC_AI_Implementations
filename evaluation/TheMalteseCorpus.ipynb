```python
import pandas as pd
```

```python
labels = ['toxicity','obscene','sexual_explicit','identity_attack','insult', 'threat','not_offensive']
```

For Train

```python
jigsaw_train = pd.read_csv('../Data/jigsaw.train.csv')#, usecols=lambda column: column != 'Unnamed: 0')
jigsaw_train.drop(inplace=True, columns=['Unnamed: 0', 'Unnamed: 0.1'])
jigsaw_train
```

```python
jigsaw_train['Maltese'] = False
jigsaw_train
```

```python
hs_train = pd.DataFrame()
for i in range(1, 4):
    if i is not 10:
        path = f'.\\data\\translated\\test_{i}.csv'
        hs_train = pd.concat([hs_train, pd.read_csv(path)],axis=0)

hs_train
```

```python
hs_train.drop(inplace=True, columns=['comment_text'])
hs_train
```

```python
hs_train.rename(columns={'Maltese': 'comment_text'}, inplace=True)
hs_train = hs_train[['comment_text','toxicity','obscene','sexual_explicit','identity_attack','insult', 'threat','not_offensive']]
hs_train
```

```python
hs_train['Maltese'] = True
```

```python
hs_train = pd.concat([hs_train, jigsaw_train], axis=0)
```

```python
hs_train = hs_train.sample(frac=1).reset_index(drop=True)
hs_train
```
